```python
data = np.genfromtxt('data.csv', delimiter=',')
```

```python
plt.plot(data[:,1])
plt.figure(figsize=(12,12), dpi=100)
plt.show()
```

```python
strong = 4*np.median(bruh)/0.6745 
strong
```

```python
ix = np.diff(np.sign(bruh-strong), axis=0) > 0
```
